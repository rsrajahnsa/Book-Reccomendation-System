# Popularity Based Reccomendation System

```python
ratings.merge(books, on='ISBN')
```

```python
ratings.merge(books, on='ISBN').shape
```

```python
ratings_with_name=ratings.merge(books, on='ISBN')
```

```python
ratings_with_name
```

```python
num_rating_df=ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
import pandas as pd
import numpy as np

# Filter out non-numeric values from the 'Book-Rating' column
ratings_with_name['Book-Rating'] = pd.to_numeric(ratings_with_name['Book-Rating'], errors='coerce')

# Drop rows with NaN values (non-numeric)
ratings_with_name.dropna(subset=['Book-Rating'], inplace=True)

# Perform aggregation using numpy.mean()
try:
    avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].agg(np.mean).reset_index()
    avg_rating_df.rename(columns={'Book-Rating': 'Avg_ratings'}, inplace=True)
except Exception as e:
    print("Error occurred during aggregation:", e)

# Check the data types and the first few rows of the resulting DataFrame
print(avg_rating_df.dtypes)
print(avg_rating_df.head())
# Why was .mean() in pandas not working here?


```

```python
avg_rating_df
```

```python
popular_df=num_rating_df.merge(avg_rating_df, on='Book-Title')
```

```python
popular_df
popular_df=popular_df[popular_df['num_ratings']>=250].sort_values('Avg_ratings',ascending=False).head(50)
```

# Collaborative Filtering Based Reccomendation System

```python
x=ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
padhe_likhe_users=x[x].index
```

```python
filtered_rating=ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
```

```python
filtered_rating
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=y[y].index
famous_books
```

```python
final_ratings=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID', values='Book-Rating')
pt.fillna(0,inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores= cosine_similarity(pt)
```

```python
def recommend(Book_Title):
    # index fetch
    index = np.where(pt.index==Book_Title)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1], reverse=True)[1:6]
    for i in similar_items:
        print(pt.index[i[0]])
    
   # return suggestions
    
```

```python
recommend('The Notebook')
```
